imports and downloads

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

uploading the data

```python
file = "C:/Users/pragy/Applied-Machine-Learning/Assignment2/smsspamcollection/SMSSpamCollection"
```

```python
data=pd.read_csv(file, sep='\t', header=None, names=['Label', 'Mssg'])
```

```python
data.head()
```

```python
print(data.info())
```

```python
data.loc[data.Label == 'spam', 'Label'] = 1
data.loc[data.Label == 'ham', 'Label'] = 0
```

```python
data.head()
```

Train test split

```python
train_X, test_X, train_Y, test_Y = train_test_split(data.Mssg, data.Label, test_size = 0.1, random_state = 42)
```

Train val split

```python
train_X, val_X, train_Y, val_Y = train_test_split(train_X, train_Y, test_size = 0.1, random_state = 42)
```

```python
## check seed 42 for data profile
print("Number of 0s and 1s in Train set are:")
print(train_Y.value_counts())

print("Number of 0s and 1s in Validation set are:")
print(val_Y.value_counts())

print("Number of 0s and 1s in Test set are:")
test_Y.value_counts()
```

```python
df_train = pd.DataFrame(list(zip(train_X,train_Y)),columns=["Text",'Label'])
df_test = pd.DataFrame(list(zip(test_X,test_Y)),columns=["Text",'Label'])
df_val = pd.DataFrame(list(zip(val_X,val_Y)),columns=["Text",'Label'])
df_val.to_csv('Data/Validation Data.csv',index=False) 
df_test.to_csv('Data/Test Data.csv',index=False) 
df_train.to_csv('Data/Training Data.csv',index=False) 
```

Repeating same for a different seed

```python
train_X, test_X, train_Y, test_Y = train_test_split(data.Mssg, data.Label, test_size = 0.1, random_state = 95)
train_X, val_X, train_Y, val_Y = train_test_split(train_X, train_Y, test_size = 0.1, random_state = 95)
```

```python
## check seed 95 for data profile
print("Number of 0s and 1s in Train set are:")
print(train_Y.value_counts())

print("Number of 0s and 1s in Validation set are:")
print(val_Y.value_counts())

print("Number of 0s and 1s in Test set are:")
test_Y.value_counts()
```

```python
df_train = pd.DataFrame(list(zip(train_X,train_Y)),columns=["Text",'Label'])
df_test = pd.DataFrame(list(zip(test_X,test_Y)),columns=["Text",'Label'])
df_val = pd.DataFrame(list(zip(val_X,val_Y)),columns=["Text",'Label'])
df_val.to_csv('Data/Validation Data.csv',index=False) 
df_test.to_csv('Data/Test Data.csv',index=False) 
df_train.to_csv('Data/Training Data.csv',index=False) 
```
